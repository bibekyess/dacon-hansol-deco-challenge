```python
import os

def list_pdf_files(directory):
    pdf_files = []
    # Iterate through all files in the directory
    for file in os.listdir(directory):
        # Check if the file extension is '.pdf'
        if file.endswith('.pdf'):
            # If it's a PDF file, add its full path to the list
            pdf_files.append(os.path.join(directory, file))
    return pdf_files

# Replace 'directory_path' with the path of your target directory
directory_path = '/content/drive/MyDrive/DACON_wallpaper_contest_data/Domain_related_references'
pdf_files_list = list_pdf_files(directory_path)
```

```python
# Global settings
from llama_index.core import Settings
# Local settings
from llama_index.core.node_parser import SentenceSplitter
import os.path
from llama_index.core import (
    VectorStoreIndex,
    SimpleDirectoryReader,
    StorageContext,
    load_index_from_storage,
)

documents = SimpleDirectoryReader(input_files=["/content/drive/MyDrive/DACON_wallpaper_contest_data/clean-files-for-retriever/good_train_long.csv"]).load_data()
documents
```

```python
# This chunk size is chosen to match the exact number of rows, the end of each entry contains the line break "\n"
# splitter = SentenceSplitter(chunk_size=462, paragraph_separator="\n") # For a question, single long answer pairs
splitter = SentenceSplitter(chunk_size=982, paragraph_separator="\n")
nodes = splitter.get_nodes_from_documents(documents)
```

```python
len(nodes)
```

```python
nodes[0]
```

```python
len(nodes+nodes_pdf_doc)
```

```python
raw_index = VectorStoreIndex(nodes=nodes+nodes_pdf_doc, insert_batch_size=2)
```

```python
PERSIST_DIR = "/content/drive/MyDrive/DACON_wallpaper_contest_data/train-plus-sup-pdf-vector-index-storage"
raw_index.storage_context.persist(persist_dir=PERSIST_DIR)
```

```python
!pip install llama-index-postprocessor-flag-embedding-reranker
!pip install git+https://github.com/FlagOpen/FlagEmbedding.git

```

```python
from llama_index.postprocessor.flag_embedding_reranker import FlagEmbeddingReranker

reranker = FlagEmbeddingReranker(
    top_n=4,
    model="Dongjin-kr/ko-reranker",
)

raw_query_engine = raw_index.as_query_engine(similarity_top_k=8, node_postprocessors=[reranker], verbose=True)

response_1 = raw_query_engine.query("방청 페인트의 종류에는 어떤 것들이 있는지 알고 계신가요? 또한, 원목사이딩을 사용하는 것에 어떤 단점이 있을까요?")
response_1.source_nodes
```
