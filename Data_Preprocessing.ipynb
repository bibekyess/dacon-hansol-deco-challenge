Return the train.csv file in the triplets of {Questions, Contexts and Answer}.

This can be used to finetune embedding and the generative model.

```python
import pandas as pd

csv_file_path = "train.csv"

train_df = pd.read_csv(csv_file_path)

train_df
```

```python
import pandas as pd
import copy

train_df_aug = train_df.copy()

# Combine the strings in 질문_1 and 질문_2 into a new 'Question' column
question_columns = ['질문_1', '질문_2']
# train_df_aug['Question'] = train_df_aug['질문_1'] + ' ' + train_df_aug['질문_2']
train_df_aug['Question'] = train_df_aug['질문_2'] + ' ' + train_df_aug['질문_1']
train_df_aug.drop(question_columns, axis=1, inplace=True)



# Find the longest text among 답변_1, 답변_2, 답변_3, 답변_4, 답변_5 and put it under 'Answer' column,
# while adding the contents of the remaining columns under 'Context' column

answer_columns = ['답변_1', '답변_2', '답변_3', '답변_4', '답변_5']

def find_longest_text(row):
    longest_text = max(row[answer_columns], key=len)
    return longest_text

# Finding the second_longest_text makes sense because in some samples, the longest text and combination of all other answers are of similar length.
# It seems like the longest text is a very good candidate for a meaningful context
# def find_second_longest_text(row):
#     sorted_answers = sorted(row[answer_columns], key=len, reverse=True)
#     second_longest_text = sorted_answers[1]
#     return second_longest_text


train_df_aug['Answer'] = train_df_aug.apply(find_longest_text, axis=1)
# train_df_aug['Answer'] = train_df_aug.apply(find_second_longest_text, axis=1)
train_df_aug['Context'] = train_df_aug.apply(lambda row: ' '.join([row[col] for col in answer_columns if row[col] != row['Answer']]), axis=1)
train_df_aug.drop(answer_columns, axis=1, inplace=True)

train_df_aug.head(2)
```
